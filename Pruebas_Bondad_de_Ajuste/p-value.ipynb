## Sobre el p-valor

```python
import scipy.stats as stats
import numpy as np
```

Del ejercicio resuelto, $\chi^{2}_{obs} \approx 7.73$, 
$$
c_{\alpha} = q_{1-\alpha}(\chi^{2}_{7})
$$

con $\alpha = 0.05$,

```python
alpha = 0.05
c_alpha = stats.chi2.ppf(q=1-alpha, df=7)
c_alpha
```

Es decir, $c_{0.05} \approx 14.06$, y como $\chi^{2}_{obs} = 7.73 \geq 14.06 = c_{0.05}$!, no rechazamos $H_0$.

El p-valor es $p_{val} = \mathbb{P}(\chi^{2}_{7} \geq \chi^{2}_{7,obs}) \approx 0.357$


```python
p_val = stats.chi2.sf(7.73, df=7)
p_val
```

El p-valor es el mayor error (del tipo I, $\alpha$) bajo el cual rechazamos $H_0$

```python
chi_obs = 7.73

for alpha in np.linspace(0.3, 0.4, 100):
    print(f"p-val = {p_val}")
    c_alpha = stats.chi2.ppf(q=1-alpha, df=7)
    if chi_obs >= c_alpha:
        print(f"alpha = {alpha}")
        print("---->Rechazamos H_0\n")
    else:
        print(f"alpha = {alpha}")
        print("No rechazamos H_0\n")
```
